```python
# 1. ติดตั้ง Library (รันครั้งเดียว)
!pip install yfinance

# 2. โค้ดดึงข้อมูลและรวมตารางอัตโนมัติ
import yfinance as yf
import pandas as pd

# กำหนดรายชื่อตัวแปรที่อยากได้ (ทอง, เงิน, น้ำมัน, หุ้น, ดอลลาร์, พันธบัตร)
tickers = {
    'Gold': 'GC=F',
    'Silver': 'SI=F',
    'Crude_Oil': 'CL=F',
    'S&P500': '^GSPC',
    'Dollar_Index': 'DX-Y.NYB',
    'Bond_10Y': '^TNX'
}

# ดึงข้อมูลย้อนหลัง (ตัวอย่าง: รายวัน ตั้งแต่ปี 2010)
df_list = []
for name, symbol in tickers.items():
    data = yf.download(symbol, start="2010-01-01", interval="1d") # 1d = รายวัน, 1mo = รายเดือน
    data = data[['Close']].rename(columns={'Close': name}) # เอาแค่ราคาปิด
    df_list.append(data)

# รวมทุกตารางเข้าด้วยกัน
df_final = pd.concat(df_list, axis=1)
df_final = df_final.dropna() # ลบวันที่ข้อมูลไม่ครบ

# ดูหน้าตาข้อมูล
print(df_final.head())

# (Optional) ถ้าอยากเซฟเก็บไว้เป็น CSV
# df_final.to_csv('gold_project_data.csv')
df_final.tail
```
