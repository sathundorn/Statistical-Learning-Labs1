### ปฏิบัติการครั้งที่ 7
1. สร้างโมเดลเพื่อการทำนายดังนี้
1.1 แบ่งข้อมูลออกเป็น 2 ส่วน
   - training set: วันที่ 1996-01-01 ถึง 2009-12-01
   - test set: วันที่ 2010-01-01 ถึง 2012-03-01  
1.2 แยกส่วนประกอบ $y_t=S_t+T_t+R_t$ บน training set
1.3 สร้าง time series ชุดใหม่ที่แสดงถึงทำนายค่าบน test set โดยนำค่า $T_t+R_t$ ของวันล่าสุดใน training set ที่มีค่า $T_t$ มาบวกกับแต่ละค่าใน $S_t$ จากวันที่ วันที่ 2010-01-01 ถึง 2012-03-01  
3. คำนวณ RMSE โดยใช้ฟังก์ชัน `rmse` ข้างล่าง
4. แสดงแผนภาพข้อมูล elecequip และค่าทำนายที่ได้

```python
ax = result_add.trend.plot(marker='o', color='blue', figsize=(20,8))
ax = result_add.observed.plot(marker='o', color='red', figsize=(20,8))
```

```python
train = data.loc['1996-01-01':'2009-12-01']
test = data.loc['2010-01-01':'2012-03-01']
```

```python
def rmse(y_true,y_pred):
  # Inputs
  # y_true: actual value, y_pred: predicted values
  # Output: RMSE between y_true and y_pred
  return np.sqrt(np.mean(np.square(y_true-y_pred)))

train = data.loc['1996-01-01':'2009-12-01']
test = data.loc['2010-01-01':'2012-03-01']

result_train = seasonal_decompose(train['value'], model='additive',period=12)
seasonal = result_train.seasonal
trend = result_train.trend
resid = result_train.resid

train_valid_trend = trend.dropna().iloc[-1]
train_valid_resid = resid.dropna().iloc[-1]
tr_last = train_valid_trend + train_valid_resid

seasonal_pattern = seasonal.groupby(seasonal.index.month).mean()
forecast = []
for idx in test.index:
    m = idx.month
    S_t = seasonal_pattern.loc[m]
    forecast.append(S_t + tr_last)

forecast = pd.Series(forecast, index=test.index, name="forecast")
rmse_val = rmse(test['value'],forecast)
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["value"], label="Actual", color="blue")
plt.plot(test.index, forecast, label="Forecast", color="red", linestyle="--")
plt.axvline(pd.to_datetime("2010-01-01"), color="black", linestyle=":")
plt.title(f"Classical Decomposition Forecast (RMSE={rmse_val:.2f})")
plt.legend()
plt.tight_layout()
```
